```python
# Analysis 1: Comparing Sample_size to Median
# Aims to answer the question
# Do students in more popular majors make more money?

recent_grads.plot(x = "Sample_size", y = "Median", kind = "scatter")
```

BACKGROUND 

The column "Sample_size" is the number of students enrolled in a particular major. For example, the major "ZOOLOGY" has a sample size of 47. This simply means that 47 Zoology majors participated in the survey. Roughly, the sample size can be a measure of popularity with respect to the majors. The more students for a particular major, the more popular it is. Conversely, the less students for a particular major, the less popular it is.


The column "Median" is the median salary of full-time, year-round workers for a particular major. Using the zoology example again, the median salary for zoology majors is 26K.


INTERPRETATION 

From the plot given above, it can be seen that more data points populate the less popular majors. It can also be seen that the median for less popular majors ranges from 20K to around 80K (excluding the outlier of a median greater than 100K).
As one moves to the left, the major becomes more popular, but the median ranges decrease. For example, sample sizes from 2000 to 3000 show a median range of slightly above 20K to slightly above 40K.

Based on the results of this survey, it would seem that students of more popular majors are less likely to make more money than those of less popular majors.

Based on the plot above, it can be seen that the data points are denser at the smaller number of full time employees, but the corresponding median salary ranges from 20K to around 80K. 


Moving to the left, there are lesser data points where those data points are more scattered and the corresponding salary reduces in its range.


In general (without considering the accurate calculation of correlation), it would seem that the data tries to converge to a median value of somewhere in between 25K and 40K. But because the data points can be scattered with respect to the Full_time column value (i.e., one full_time value can have multiple possible Median value), this link can be considered weak.

```python
# Analysis 9: Histogram for Median
# Aims to answer the question:
# What's the most common median salary range?

recent_grads["Median"].hist(bins = 40, range = (0, 40000))
```

```python
# Determining the most common salary range

n = 36
median_interval = 40000 / 40

max_range = n*median_interval
min_range = max_range - median_interval

print("MOST COMMON MEDIAN SALARY RANGE")
print("(", min_range, ", ", max_range, ")")
```

The histogram for the median was created using 40 bins with a range from 0 to 40K. From the histogram, it was found that the highest bar was the 36th bar (n = 36). The interval for each range is 40K / 40 (highest value for the range divided by the number of bins). Thus, the range for the highest bar is calculated as:

maximum range = n X interval

minimum range = maximum range - interval

This is the formula used for the most common median salary range calculation.

```python
# Calculating percentages - based on the histogram

total = 75 + 95
male_dom = (75 / total)*100
female_dom = (95 / total)*100

print("PERCENT DOMINATION BASED ON HISTOGRAM")
print("male: ", male_dom, "%")
print("female: ", female_dom, "%")

# Calculating percentages - based on actual count
gender_domination = recent_grads["ShareWomen"]
male_dom_bool = gender_domination < 0.5
male_dom_series = gender_domination[male_dom_bool]
female_dom_series = gender_domination[male_dom_bool == False]

male_dom_total = len(male_dom_series)
female_dom_total = len(female_dom_series)

male_dom_percent_actual = (male_dom_total / cleaned_data_count)*100
female_dom_percent_actual = (female_dom_total / cleaned_data_count)*100

print()
print("PERCENT DOMINATION BASED ON ACTUAL COUNT")
print("male: ", male_dom_percent_actual, "%")
print("female: ", female_dom_percent_actual, "%")
```

BACKGROUND

Using Zoology as an example, the number of women majoring in Zoology is 5359 while the total number of students majoring in Zoology is 8409. Dividing 5359 by 8409, we get 0.63729. This ratio is the ShareWomen for Zoology.

DEFINITIONS

Predominantly Female => Sharewomen with a value of 0.5 to 1

Predominantly Male => Sharewomen with a value of 0 to 0.49

INTERPRETATION

From the histogram for ShareWomen, the range from 0 to 0.49 has a frequency of approximatly 75 while the Sharewomen from 0.5 to 1 has a frequency of approximately 95. Based on Histogram approximations, the percent of majors predominantly male is 44.12% while the percent majors predominantly female is 55.88%.

Actual values may be calculated through manipulations of the column ShareWomen. The actual values were found to be close to the approximation - 44.19% for males and 55.81% for females.

```python
# Analysis 20 - Suggested Next Steps
# Use a grouped bar plot to compare the number of men with women in each category of majors

# Grouping and summing the number of men and women per category
# End Result: Lists for category titles, number of men per category, number of women per category
category_list = recent_grads["Major_category"].unique().tolist()

def category_series(col_title, mother_data = recent_grads, categories = category_list):
    series_column = mother_data[col_title]
    col_per_category = []
    
    for cat in categories:
        bool = mother_data["Major_category"] == cat
        subset_column = series_column[bool]
        total = subset_column.sum()
        col_per_category.append(total)
    
    #series_col_per_category = pd.Series(data = col_per_category, index = categories)
    return(col_per_category)


men_per_category_list = category_series("Men")
women_per_category_list = category_series("Women")

# Creating a new dataframe
# End Result: A dataframe with columns for Categories, Number of men per category and Number of women per category
new_column_titles = ["Categories", "Men", "Women"]
bundled_data_list = [category_list, men_per_category_list, women_per_category_list]

dictionary_by_category = {}

for i, title in enumerate(new_column_titles):
    dictionary_by_category[title] = bundled_data_list[i]


df_by_category = pd.DataFrame(dictionary_by_category)


# Plotting the Data Set
# End Result: A grouped bar plot comparing the number of men and number of women per category
df_by_category.plot(x = "Categories", kind = "bar")
```

From the plot above, it can be seen that the most number of students are enrolled in Business, but the the number of male students are greater than the number of female students.

Judging by the very low heights of the bars, there are only a few students enrolled in interdisciplinary courses but the number of female students are greater than the number of male students.
